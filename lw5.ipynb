```python
import numpy as np
import matplotlib.pyplot as plt

# constants
xk = [0.11, 0.15, 0.21, 0.29, 0.35, 0.40]  # x coordinates
fk = [9.05421, 6.61659, 4.69170, 3.35106, 2.73951, 2.36522]  # y coordinates
xs = [0.186, 0.275, 0.332]  # control points


def polynom_of_lagrange(vx, vy):
    def lagrange(x):
        g = 0
        for j in range(len(vy)):
            # basis polynomial
            d1 = 1
            d2 = 1
            for i in range(len(vx)):
                if i != j:
                    d1 = d1 * (x - vx[i])
                    d2 = d2 * (vx[j] - vx[i])
            g = g + vy[j] * d1 / d2
        return g

    return lagrange


lagrange = polynom_of_lagrange(xk, fk)

print("Наближені значення функції у контрольних точках за Лагранжем:")
fs = [lagrange(x) for x in xs]
print(
    f"lagrange({xs[0]})={fs[0]:.7f}, lagrange({xs[1]})={fs[1]:.7f}, lagrange({xs[2]})={fs[2]:.7f}"
)

vx = np.linspace(np.min(xk), np.max(xk), 100)
vy = [lagrange(x) for x in vx]

plt.figure(figsize=(10, 6))
plt.plot(vx, vy, color="blue", linewidth=2.5, label="Лагранжа")
plt.plot(
    xk,
    fk,
    "o",
    color="yellow",
    markeredgecolor="black",
    markersize=8,
    label="табличні значення",
)
plt.plot(
    xs,
    fs,
    "D",
    color="red",
    markeredgecolor="black",
    markersize=6,
    label="значення в контрольних точках",
)
plt.grid(True, alpha=0.7)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

print("\nПобудова інтерполяційного поліному Ньютона:")

print("1) f(0.11) = 9.05421")
p0 = lambda x: a0
a0 = 9.05421
print(f"1) p0(0.11)={p0(0.11)}")

print("\n2) f(0.15) = 6.61659")
p1 = lambda x: p0(x) + a1 * (x - 0.11)
a1 = (6.61659 - p0(0.15)) / (0.15 - 0.11)
print(f"2) p1(0.11)={p1(0.11)}, p1(0.15)={p1(0.15)}")

print("\n3) f(0.21) = 4.69170")
p2 = lambda x: p1(x) + a2 * (x - 0.11) * (x - 0.15)
a2 = (4.69170 - p1(0.21)) / ((0.21 - 0.11) * (0.21 - 0.15))
print(f"3) p2(0.11)={p2(0.11)}, p2(0.15)={p2(0.15)}, p2(0.21)={p2(0.21)}")

print("\n4) f(0.29) = 3.35106")
p3 = lambda x: p2(x) + a3 * (x - 0.11) * (x - 0.15) * (x - 0.21)
a3 = (3.35106 - p2(0.29)) / ((0.29 - 0.11) * (0.29 - 0.15) * (0.29 - 0.21))
print(
    f"4) p3(0.11)={p3(0.11)}, p3(0.15)={p3(0.15)}, p3(0.21)={p3(0.21)}, p3(0.29)={p3(0.29)}"
)

print("\n5) f(0.35) = 2.73951")
p4 = lambda x: p3(x) + a4 * (x - 0.11) * (x - 0.15) * (x - 0.21) * (x - 0.29)
a4 = (2.73951 - p3(0.35)) / (
    (0.35 - 0.11) * (0.35 - 0.15) * (0.35 - 0.21) * (0.35 - 0.29)
)
print(
    f"5) p4(0.11)={p4(0.11)}, p4(0.15)={p4(0.15)}, p4(0.21)={p4(0.21)}, p4(0.29)={p4(0.29)}, p4(0.35)={p4(0.35)}"
)

print("\n6) f(0.40) = 2.36522")
p5 = lambda x: p4(x) + a5 * (x - 0.11) * (x - 0.15) * (x - 0.21) * (
    x - 0.29
) * (x - 0.35)
a5 = (2.36522 - p4(0.40)) / (
    (0.40 - 0.11)
    * (0.40 - 0.15)
    * (0.40 - 0.21)
    * (0.40 - 0.29)
    * (0.40 - 0.35)
)
print(
    f"6) p5(0.11)={p5(0.11)}, p5(0.15)={p5(0.15)}, p5(0.21)={p5(0.21)}, p5(0.29)={p5(0.29)}, p5(0.35)={p5(0.35)}, p5(0.40)={p5(0.40)}"
)

# compare with lagrange
print(
    "\nНаближені значення функції у контрольних точках за Ньютоном та Лагранжем:"
)
newton_values = [p5(x) for x in xs]
print(
    f"newton({xs[0]})={newton_values[0]:.7f}, newton({xs[1]})={newton_values[1]:.7f}, newton({xs[2]})={newton_values[2]:.7f}"
)
print(
    f"lagrange({xs[0]})={fs[0]:.7f}, lagrange({xs[1]})={fs[1]:.7f}, lagrange({xs[2]})={fs[2]:.7f}"
)

vx = np.linspace(np.min(xk), np.max(xk), 100)
vy_lagrange = [lagrange(x) for x in vx]
vy_newton = [p5(x) for x in vx]

plt.figure(figsize=(10, 6))
plt.plot(vx, vy_lagrange, color="blue", linewidth=2.5, label="Лагранжа")
plt.plot(
    vx,
    vy_newton,
    color="green",
    linestyle="--",
    linewidth=2.5,
    label="Ньютона",
)
plt.plot(
    xk,
    fk,
    "o",
    color="yellow",
    markeredgecolor="black",
    markersize=8,
    label="табличні значення",
)
plt.plot(
    xs,
    fs,
    "D",
    color="red",
    markeredgecolor="black",
    markersize=6,
    label="контрольні точки",
)
plt.grid(True, alpha=0.7)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

"""
Висновок: Обидва методи дають однакові результати, оскільки теоретично 
вони обчислюють один і той же інтерполяційний поліном, але різними способами.
"""
```
